## Methodology 
#### The core of this analysis getting around the main problem that the census internet subscription data does not exist on a tract level of detail. The most granular level of detail is on the county level, and that just isn't sufficient for this analysis. There does, however, exist income data on a census tract level. And, internet subscription data does breakdown some of it's data by income. 

#### So I used income data to link the data sources together. I'm assumeing and applying the county level internet rates by income to the income data of each census tract.
